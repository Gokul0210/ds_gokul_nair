```python
# 1. Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)

# 2. Load datasets

trader_df = pd.read_csv('/content/historical_data.csv')
sentiment_df = pd.read_csv('/content/fear_greed_index.csv')

print("Trader Data Preview:")
display(trader_df.head())

print("\nSentiment Data Preview:")
display(sentiment_df.head())

# 3. Inspect columns

print("\nTrader Data Columns:")
print(trader_df.columns.tolist())

print("\nSentiment Data Columns:")
print(sentiment_df.columns.tolist())

# 4. Convert sentiment date

sentiment_df['date'] = pd.to_datetime(sentiment_df['date'], errors='coerce')

# 5. Convert trader timestamp (Timestamp IST)

trader_df['time'] = pd.to_datetime(
    trader_df['Timestamp IST'],
    format='%d-%m-%Y %H:%M',
    errors='coerce'
)

# Create date column for merging
trader_df['date'] = pd.to_datetime(trader_df['time'].dt.date)

# Verify conversion
print("\nTrader date preview:")
display(trader_df[['Timestamp IST', 'time', 'date']].head())

# 6. Merge trader data with sentiment

merged_df = trader_df.merge(
    sentiment_df[['date', 'classification']],
    on='date',
    how='left'
)

print("\nMerged Data Preview:")
display(merged_df[['Account', 'date', 'classification', 'Closed PnL']].head())

# 7. Prepare analysis dataset

analysis_df = merged_df[[
    'Account',
    'Coin',
    'Side',
    'Size USD',
    'Closed PnL',
    'Fee',
    'classification'
]].copy()

# Handle missing sentiment
analysis_df['classification'] = analysis_df['classification'].fillna('Unknown')

analysis_df.head()

# 8. Trader-level performance metrics

analysis_df['win'] = analysis_df['Closed PnL'] > 0

trader_summary = analysis_df.groupby(
    ['Account', 'classification']
).agg(
    total_pnl=('Closed PnL', 'sum'),
    avg_trade_size=('Size USD', 'mean'),
    trade_count=('Account', 'count'),
    win_rate=('win', 'mean'),
    total_fees=('Fee', 'sum')
).reset_index()

trader_summary['win_rate'] = trader_summary['win_rate'] * 100

print("\nTrader Summary Preview:")
display(trader_summary.head())

contrarian_traders = trader_summary[
    trader_summary['classification'].isin(['Fear', 'Extreme Fear'])
].sort_values(
    by='total_pnl',
    ascending=False
).groupby('classification').head(10)

contrarian_traders

# 9. Save processed CSV (MANDATORY OUTPUT)

trader_summary.to_csv(
    'processed_trader_sentiment_summary.csv',
    index=False
)

print("Processed CSV saved: processed_trader_sentiment_summary.csv")

# 10. Aggregate insights (Fear vs Greed)

sentiment_summary = trader_summary.groupby('classification').agg(
    avg_total_pnl=('total_pnl', 'mean'),
    avg_win_rate=('win_rate', 'mean'),
    avg_trade_count=('trade_count', 'mean')
).reset_index()

print("\nSentiment-level Summary:")
display(sentiment_summary)

# 11. Visualization: PnL by Sentiment

plt.figure(figsize=(8, 5))
sns.boxplot(
    data=analysis_df,
    x='classification',
    y='Closed PnL'
)
plt.title('Closed PnL Distribution by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Closed PnL')
plt.tight_layout()
plt.savefig('pnl_by_sentiment.png')
plt.show()

print("Plot saved: pnl_by_sentiment.png")

# 12. Sanity info

print("\nTrader Data Info:")
trader_df.info()

print("\nSentiment Data Info:")
sentiment_df.info()
```
